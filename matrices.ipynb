```python
def spiral_matrices(matrix):
    if not matrix or not matrix[0]:
        return []

    top, bottom = 0, len(matrix) - 1
    left, right = 0, len(matrix[0]) - 1
    result = []

    while left <= right and top <= bottom:
        # Traverse top row
        for i in range(left, right + 1):
            result.append(matrix[top][i])
        top += 1

        # Traverse right column
        for i in range(top, bottom + 1):
            result.append(matrix[i][right])
        right -= 1

        # Traverse bottom row (if still within bounds)
        if top <= bottom:
            for i in range(right, left - 1, -1):
                result.append(matrix[bottom][i])
            bottom -= 1

        # Traverse left column (if still within bounds)
        if left <= right:
            for i in range(bottom, top - 1, -1):
                result.append(matrix[i][left])
            left += 1

    return result
matrix = [
    [ 1,  2,  3,  4,  5,  6,  7,  8],
    [48, 49, 50, 51, 52, 53, 54,  9],
    [47, 64, 65, 66, 67, 68, 55, 10],
    [46, 63, 62, 61, 60, 69, 56, 11],
    [45, 44, 43, 42, 41, 70, 57, 12],
    [40, 39, 38, 37, 36, 35, 34, 13]
]
print(spiral_matrices(matrix))
```

```python
from typing import List

class Solution:
    def isToeplitzMatrix(self, matrix: List[List[int]]) -> bool:
        m, n = len(matrix), len(matrix[0])

        # Check diagonals
        for i in range(1, m):
            for j in range(1, n):
                if matrix[i][j] != matrix[i - 1][j - 1]:
                    return False
        return True

matrix = [
    [1,2,3,4],
    [5,1,2,3],
    [9,5,1,2]
]

print(Solution().isToeplitzMatrix(matrix))

```
