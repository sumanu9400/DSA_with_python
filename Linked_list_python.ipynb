# Linked list

```python
#single linked list
class Node:
  def  __init__(self,data):
    self.data = data
    self.next = None

class LinkedList:
  def __init__(self):
    self.head = None

  def print_linked_list(self):
    curr = self.head
    while(curr!=None):
      print(curr.data)
      curr = curr.next

def main():
  ll = LinkedList()

  p1 = Node(10)
  p2 = Node(20)
  p3 = Node(30)
  p4 = Node(40)

  ll.head = p1
  p1.next = p2
  p2.next = p3
  p3.next = p4

  ll.print_linked_list()

if __name__=='__main__':
  main()
```

Double linked list


```python
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None

class DoublyLinkedList:
    def __init__(self):
        self.head = None

    # ðŸ”¸ Insert at the End
    def append(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new_node
        new_node.prev = temp
```

Circular linked list


```python
class Node:
  def __init__ (self,data):
    self.data = data
    self.next = None

class CircularLinkedList():
  def __init__(self):
    self.head = None

  def print_list(self):
    if not self.head:
      print("list an empty")
      return

    curr = self.head
    while True :
      print(curr.data)
      curr = curr.next
      if curr == self.head:
        break


  def delete_at_begin(self):
      if self.head is None:
          return  # Empty list

      # Case: Only one node
      if self.head.next == self.head:
          self.head = None
          return

      # Case: More than one node
      last = self.head
      while last.next != self.head:
          last = last.next

      # Move head to next node
      last.next = self.head.next
      self.head = self.head.next

def main():
    cll = CircularLinkedList()

    # Creating nodes
    p1 = Node(10)
    p2 = Node(20)
    p3 = Node(30)
    p4 = Node(40)

    # Linking circularly
    cll.head = p1
    p1.next = p2
    p2.next = p3
    p3.next = p4
    p4.next = p1  # Circular link

    print("Before deletion:")
    cll.print_list()

    cll.delete_at_begin()

    print("\nAfter deletion:")
    cll.print_list()

if __name__ == '__main__':
    main()
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None  # For doubly linked list

class LinkedList:
    def __init__(self):
        self.head = None

    def print_linked_list(self):
        curr = self.head
        while curr is not None:
            print(curr.data, end="->")
            curr = curr.next
        print()

    def delete_at_index(self, index):
        if self.head is None:
            print("List is empty")
            return

        curr = self.head
        count = 0

        # If head needs to be removed
        if index == 0:
            self.head = curr.next
            if self.head is not None:
                self.head.prev = None
            return

        # Traverse to the node to be deleted
        while curr is not None and count < index:
            curr = curr.next
            count += 1

        if curr is None:
            print("Index out of range")
            return

        # Remove curr node
        if curr.next is not None:
            curr.next.prev = curr.prev

        if curr.prev is not None:
            curr.prev.next = curr.next

def main():
    ll = LinkedList()

    p1 = Node(10)
    p2 = Node(20)
    p3 = Node(30)
    p4 = Node(40)

    ll.head = p1
    p1.next = p2
    p2.prev = p1
    p2.next = p3
    p3.prev = p2
    p3.next = p4
    p4.prev = p3

    print("Original list:")
    ll.print_linked_list()

    print("Delete node at index 2:")
    ll.delete_at_index(2)  # This should delete node with value 30

    ll.print_linked_list()

if __name__ == '__main__':
    main()
```
